```python
#A1

import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
import IPython.display as ipd

y, sr = librosa.load(r'C:\Users\deven\Downloads\Recording (3).wav')
first_derivative = np.diff(y)
first_derivative /= np.max(np.abs(first_derivative))

print("First Derivative Signal:")

plt.figure(figsize=(10, 5))
librosa.display.waveshow(first_derivative, sr=sr,color='g')
plt.title('Segment Waveform')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.tight_layout()
plt.show()

ipd.Audio(first_derivative, rate=sr)
```
